```python
class Statistics:
    def __init__(self, datax):
        '''take arguement to initialize the instance of the object'''
        self.raw = datax
        self.dump = []
        self.data = []
        self.clean()

    def clean(self):
      for num in self.raw:
        if type(num) is int:
           self.data.append(num)
        else:
           self.dump.append(num)
           
    def clean_data(self):
       return self.data
    
    def dump_data(self):
       return self.dump

    def count(self):
        return len(self.data)
    
    def sum(self):
        _sum = 0
        for i in self.data:
            _sum += i
        return _sum 

    def max(self):
        maxnum = None
        for num in self.data:
            if maxnum == None or num > maxnum:
                maxnum = num
        return maxnum

    def min(self):
        minnum = self.data[0]
        for num in self.data:
            if  minnum > num:
                minnum = num
        return minnum

    def mean(self, dec=0):
        count = self.count()
        data_sum = 0
        for i in self.data:
            data_sum += i
        meanv = data_sum/count
        return round(meanv,dec)
    
    def median(self):
        record = self.data.copy()
        record.sort() 
        count = self.count()
        middle = None
        if count%2 == 0:
            mid = count//2
            middle = (record[mid-1] + record[mid])/2
        else:
            mid = count//2
            middle = record[mid]
        return middle
    
    def mode(self):
        '''This function is to calculate mode of a list of values'''
        mode = None
        uni_num = set(self.data)
        mcount = 0
        for x in uni_num:
            if self.data.count(x) > mcount:
                mode = x
                mcount = self.data.count(x) 
        return (mode,mcount)
    
    def range(self):
        '''This function is to calculate the rAnge of a list of values'''
        anum = self.data[0]
        znum = anum
        num_set = set(self.data)
        for i in num_set:
            if i > znum:
                znum = i
            elif i < anum:
                anum = i
        return abs(znum - anum)
    
    def var(self, dec=1):
        '''This function calculates the variance of a list of values'''
        v_mean = self.mean()
        count = self.count()
        v_sum = 0
        for x in self.data:
            v_sum += (x - v_mean) ** 2 
        return round(v_sum/count, dec)

    def std(self, dec=2):
        '''This function calculates standard deviation of list of vlaues'''
        s_deviation = self.var(dec)
        return round(s_deviation ** 0.5, dec)
    
    def freq_dist(self, dec=2):
        ''' This function serves to calculate the percentile from its frequency'''
        _item = self.data.copy()
        _item.sort()
        item = set(_item)
        freq = []
        count = 0
        tcount = self.count()
        for j in item:
            count = self.data.count(j)
            _perc = round((count/tcount)*100, dec)
            freq.append((_perc, j))
        freq.sort(reverse=True)
        return freq
    
    def describe(self):
        ''' Gives back all the statistics result'''
        funcN = ['count: ', 'sum: ', 'min: ', 'max: ', 'Range: ', 'Mean: ', 'Median: ', 'Mode: ', 'Standard Deviation: ',  'Variance: ', 'Frequency Distribution: ']
        func = [self.count(), self.sum(), self.min(), self.max(), self.range(), self.mean(), self.median(), self.mode(), self.std(), self.var(), self.freq_dist()]
        prt = ''
        for x in range(11):
            prt += funcN[x]
            prt += str(func[x])
            prt += '\n'
        return prt
        
```

```python
ages = [31, 26, 34, 37, 27, 26, 32, 32, 26, 27, 27, 24, 32, 33, 27, 25, 26, 38, 37, 31, 34, 24, 33, 29, 26]
stat = Statistics(ages)
print(stat.describe())
```

```python
class PersonAccount:

    def __init__(self, firstname, lastname, income='Null', expenses='Null'):
        self.first = firstname
        self.lastname = lastname
        self.incomes = set()
        self.expenses = set()
        try:
            self.incomes.add(income)
        except NameError:
            print('Input valid income')
        except TypeError:
            print('Enter valid income')
        
        try:
            self.expenses.add(expenses)
        except NameError:
            print('Input valid income')
        except TypeError:
            print('Enter valid income')

        self.asset = [0]
        self.withdraw = [0]

    def check(self, value):
        _value = value
        try:
            income = int(_value)
            return abs(income)
        except TypeError:
            pass

    def total_income(self):
        asset = 0
        for value in self.asset:
            asset += value
        return asset
        
    def total_expense(self):
        amount = 0
        for value in self.withdraw:
            amount += value
        return amount
    
    def account_info(self):
        details = ['Account Name: ', 'Account Number: ', 'Account Balance: ', 'Depoit: ', 'Expeneses: ']
        info = [f'{self.firstname} {self.lastname}', '001', self.account_balance(), self.total_income(), self.total_expense()]
        mark = ''
        for x in range(5):
            mark += details[x]
            mark += str(info[x])
            mark += '\n'


    def add_income(self, amount):
        self.asset.append(self.check(amount))

    def add_expense(self, amount):
        self.withdraw.append(self.check(amount))

    def account_balance(self):
        inc = self.total_income
        exp = self.total_expense
        balance =  inc - exp 
        return balance

```

```python
my = PersonAccount('James', 'Williams', 'Business')
print(my.account_balance())
```
